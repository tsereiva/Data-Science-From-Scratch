```python
#List all sheets in excel
sheets = ['hRH1-WT Prevention', 'hRH1-WT PladB 24HR', 'hRH1-WT DOX 48HR',
          'hRH1-MUT DOX 48HR', 'hRH1-MUT PladB 24HR', 'hRH1-MUT Prevention', 'mCherry DOX 48HR',
          'mCherry PladB 24HR', 'mCherry Prevention']
```

```python
#Read in each sheet and store in dictionary
sheets_dict={}
for i in sheets:
    sheets_dict[i]= pd.read_excel("C:\\Users\\tadas\\OneDrive\\Desktop\\Cell Culture\\U2OS hRH1-WT-MUT-mCherry PladB Damage\\U2OS hRH1-WT-MUT-mCherry PladB Damage and Prevention.xlsx",i)
```

```python
for i in sheets:
    sheets_dict[i] = pd.DataFrame(sheets_dict[i])

sheets_dict
```

```python
#extract DAPI, FLAG, mCherry, and Foci data from individual sheets for scatterplots
DAPI = pd.DataFrame()
FLAG = pd.DataFrame()
mCherry = pd.DataFrame()
Foci = pd.DataFrame()
Nuc_area = pd.DataFrame()
gH2AX = pd.DataFrame()

for i in sheets:
    x = sheets_dict[i]
    dc = x['DAPI CTCF']
    if 'FLAG CTCF' in x:
        fc = x['FLAG CTCF']
        FLAG[i] = fc
    if 'mCherry CTCF' in x:
        mc = x['mCherry CTCF']
        mCherry[i] = mc
    foci = x['Foci']
    na = x.iloc[:,2]
    gh2 = x['gH2AX CTCF']
    gH2AX[i] = gh2
    DAPI[i] = dc
    Foci[i] = foci
    Nuc_area[i] = na

gH2AX
```

```python
#Remove PladB Samples from from Foci and FLAG sheets for scatterplot construction, axis=1 specifies column deletion
col_drop = ["hRH1-WT PladB 24HR", "hRH1-MUT PladB 24HR"]
Foci_noPladB = Foci.drop(col_drop, axis = 1)
FLAG_noPladB = FLAG.drop(col_drop, axis = 1)
gH2AX_noPladB = gH2AX.drop(col_drop, axis = 1)
mCherry_noPladB = mCherry.drop("mCherry PladB 24HR", axis = 1).dropna()

Foci_noPladB_l= pd.melt(Foci_noPladB)
FLAG_noPladB_l = pd.melt(FLAG_noPladB)
mCherry_noPladB_l = pd.melt(mCherry_noPladB)
gH2AX_noPladB_l = pd.melt(gH2AX_noPladB)

FLAG_noPladB_l_r = FLAG_noPladB_l.rename(columns={'value': 'FLAG'})
Foci_noPladB_l_r = Foci_noPladB_l.rename(columns={'value': 'Foci'})
mCherry_noPladB_l_r = mCherry_noPladB_l.rename(columns={'value': 'mCherry'})
gH2AX_noPladB_l_r = gH2AX_noPladB_l.rename(columns={'value':'gH2AX'})

```

```python
#Subsetting mCherry only Foci
mCherry_noPladB_samples = ["mCherry DOX 48HR", "mCherry Prevention"]

mCherry_Foci = Foci_noPladB[mCherry_noPladB_samples].dropna()
mCherry_gH2AX = gH2AX_noPladB[mCherry_noPladB_samples].dropna()


mCherry_Foci_l = pd.melt(mCherry_Foci)
mCherry_Foci_l_r = mCherry_Foci_l.rename(columns={'value': 'Foci'})

mCherry_gH2AX_l = pd.melt(mCherry_gH2AX)
mCherry_gH2Ax_l_r = mCherry_gH2AX_l.rename(columns={'value': 'gH2AX'})


#Merging mCherry Foci with mCherry intensity
Foci_gH2AX_mCherry_merge = pd.concat([mCherry_Foci_l_r, mCherry_noPladB_l_r,mCherry_gH2Ax_l_r], axis=1, join = 'inner')



with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(Foci_gH2AX_mCherry_merge)
```

```python
df_t_o_Foci = df_t.dropna(subset='Foci')
#reindex due to dropped rows
df_t_o_Foci = df_t_o_Foci.reset_index(drop=True)


def outlier_removal(dfc, col):
    # Calculate the upper and lower limits
    Q1 = dfc[col].quantile(0.25)
    Q3 = dfc[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR

    # Above Upper bound
    upper = Q3+1.5*IQR
    upper_array = np.array(dfc[col] >= upper)
    print("Upper Bound:", upper)
    print(upper_array.sum())

    # Below Lower bound
    lower = Q1-1.5*IQR
    lower_array = np.array(dfc[col] <= lower)
    print("Lower Bound:", lower)
    print(lower_array.sum())

    # Create arrays of Boolean values indicating the outlier rows
    upper_array = np.where(dfc[col] >= upper)[0]
    lower_array = np.where(dfc[col] <= lower)[0]

    ##!!!! Note: inplace=True modifies input original df, subsesquent
    ##reruns of this function on the same df will result in outlier
    ##elimination each round
    dfc.drop(index=upper_array, inplace=True)
    dfc.drop(index=lower_array, inplace=True)

    return dfc

outlier_removal(df_t_o_Foci, 'Foci')
```

```python
#Cleaning up dataframes from HeLa hRH1-WT/MUT/mCherry 53bp1

#Remove PladB Samples from from Foci, FLAG, and mCherry sheets for scatterplot construction, axis=1 specifies column deletion
Hela_53bp1_col_drop = ["hRH1-WT PladB 24HR", "hRH1-MUT PladB 24HR"]
Foci_noPladB_HeLa_53bp1 = Foci_Hela_53bp1.drop(Hela_53bp1_col_drop, axis = 1)
FLAG_noPladB_HeLa_53bp1 = FLAG_Hela_53bp1.drop(Hela_53bp1_col_drop, axis = 1)
mCherry_noPladB_HeLa_53bp1 = mCherry_Hela_53bp1.drop("mCherry PladB 24HR", axis = 1).dropna()

Foci_noPladB_l_HeLa_53bp1= pd.melt(Foci_noPladB_HeLa_53bp1)
FLAG_noPladB_l_HeLa_53bp1 = pd.melt(FLAG_noPladB_HeLa_53bp1)
mCherry_noPladB_l_HeLa_53bp1 = pd.melt(mCherry_noPladB_HeLa_53bp1)

Foci_noPladB_l_HeLa_53bp1_r = Foci_noPladB_l_HeLa_53bp1.rename(columns={'value': 'Foci'})
FLAG_noPladB_l_HeLa_53bp1_r = FLAG_noPladB_l_HeLa_53bp1.rename(columns={'value': 'FLAG'})
mCherry_noPladB_l_HeLa_53bp1_r = mCherry_noPladB_l_HeLa_53bp1.rename(columns={'value': 'mCherry'})


#Subsetting mCherry only Foci
mCherry_noPladB_samples_HeLa_53bp1 = ["mCherry DOX 48HR", "mCherry Prevention"]

mCherry_Foci_HeLa_53bp1 = Foci_noPladB_HeLa_53bp1[mCherry_noPladB_samples_HeLa_53bp1].dropna()

mCherry_Foci_l_HeLa_53bp1 = pd.melt(mCherry_Foci_HeLa_53bp1)
mCherry_Foci_l_HeLa_53bp1_r = mCherry_Foci_l_HeLa_53bp1.rename(columns={'value': 'Foci'})


#Merging mCherry Foci with mCherry intensity
Foci_mCherry_merge_HeLa_53bp1 = pd.concat([mCherry_Foci_l_HeLa_53bp1_r, mCherry_noPladB_l_HeLa_53bp1_r], axis=1, join = 'inner')

FLAG_Foci_merge_HeLa_53bp1 = pd.concat([Foci_noPladB_l_HeLa_53bp1_r, FLAG_noPladB_l_HeLa_53bp1_r], axis=1, join="inner")

##Create new column with treatment type for each sample for later graphs
def treatment_type(row):
   if 'DOX' in str(row['variable']):
      return 'DOX'
   elif 'Prevention' in str(row['variable']):
      return 'Prevention'

FLAG_Foci_merge_HeLa_53bp1['treatment'] = FLAG_Foci_merge_HeLa_53bp1.apply(treatment_type, axis=1)
Foci_mCherry_merge_HeLa_53bp1['treatment'] = Foci_mCherry_merge_HeLa_53bp1.apply(treatment_type, axis=1)

Foci_mCherry_merge_HeLa_53bp1['type'] = 'cherry'
Foci_mCherry_merge_HeLa_53bp1['intensity'] = Foci_mCherry_merge_HeLa_53bp1['mCherry']

FLAG_Foci_merge_HeLa_53bp1['type'] = 'FLAG'
FLAG_Foci_merge_HeLa_53bp1['intensity'] = FLAG_Foci_merge_HeLa_53bp1['FLAG']

t_1_HeLa_53bp1 = FLAG_Foci_merge_HeLa_53bp1[['variable', 'Foci', 'type', 'intensity', 'treatment']]
t_2_HeLa_53bp1 = Foci_mCherry_merge_HeLa_53bp1[['variable', 'Foci', 'type', 'intensity','treatment']]

df_t_HeLa_53bp1 = pd.concat([t_1_HeLa_53bp1, t_2_HeLa_53bp1]).dropna().reset_index(drop=True)

outlier_removal(df_t_HeLa_53bp1, 'Foci')
```
